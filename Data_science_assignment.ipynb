Utilize the sales data set. The sales data contains transactional sales information for each sales person. It also contains the date of sales, item sold, price of each item, sales amount, region and their corresponding manager information.

```python
import numpy as np
import pandas as pd

df = pd.read_excel('SalesData.xlsx')
df
```

1. Find the least amount sale that was done for each item.

```python
least_amt_sale = df.groupby('Item')['Sale_amt'].min().reset_index()
least_amt_sale
```

2. Compute the total sales for each year and region across all items

```python
df['Year'] = df['OrderDate'].dt.year
total_sales_each_year = df.groupby(['Year', 'Region'])['Sale_amt'].sum().reset_index()
total_sales_each_year
```

3. Create new column 'days_diff' with number of days difference between reference date passed and each order date

```python
reference_date = pd.to_datetime('19-01-06', format='%y-%m-%d')
df['days_diff'] = (reference_date - df['OrderDate']).dt.days
df
```

4. Create a dataframe with two columns: 'manager', 'list_of_salesmen'. Column 'manager' will contain the unique managers present and column 'list_of_salesmen' will contain an array of all salesmen under each manager.

```python
manager_salesmen = df.groupby('Manager')['SalesMan'].unique().reset_index()
manager_salesmen.columns = ['Manager', 'List_of_Salesmen']
manager_salesmen
```

5. For all regions find number of salesman and total sales. Return as a dataframe with three columns - Region, salesmen_count and total_sales

```python
region_salesmen_sales = df.groupby('Region').agg({'SalesMan': 'nunique', 'Sale_amt': 'sum'}).reset_index()
region_salesmen_sales.columns = ['Region', 'Salesmen_count', 'Total_sales']
region_salesmen_sales
```

 6. Create a dataframe with total sales as percentage for each manager. Dataframe to contain manager and percent_sales

```python
total_sales = df['Sale_amt'].sum()
manager_percent_sales = df.groupby('Manager')['Sale_amt'].sum().reset_index()
manager_percent_sales['Percent_sales'] = (manager_percent_sales['Sale_amt'] / total_sales) * 100
manager_percent_sales
```

 Utilize the diamonds data set. The diamonds data set contains the various dimensions and information for each diamond.

```python
df3 = pd.read_csv('diamonds.csv')
df3
```

11. Count the duplicate rows of diamonds DataFrame.

```python
duplicate_rows_count = df3.duplicated().sum()
duplicate_rows_count
```

12. Drop rows in case of missing values in carat and cut columns.

```python
df3_dropped = df3.dropna(subset=['carat', 'cut'])
df3_dropped
```

```python
print("The shape before dropping : ",df3.shape)
print("The shape after dropping : ",df3_dropped.shape)
```

 13. Subset the dataframe with only numeric columns.

```python
numeric_df3 = df3_dropped.select_dtypes(include=['number'])
numeric_df3
```

14. Compute volume as (xyz) when depth is greater than 60. In case of depth less than 60 default volume to 8.

```python
df3_dropped.loc[df3_dropped['depth'] > 60, 'volume'] = df3_dropped['x'] * df3_dropped['y'] * df3_dropped['z']
df3_dropped.loc[df3_dropped['depth'] <= 60, 'volume'] = 8
df3_dropped
```

15. Impute missing price values with mean.

```python
mean_price = df3_dropped['price'].mean()
df3_dropped.loc[df3_dropped['price'].isnull(), 'price'] = mean_price
df3_dropped
```
